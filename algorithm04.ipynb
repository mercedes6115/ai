```python
# 정렬 알고리즘 sorting algorithm

# 선택정렬 (단순 선택법) Insertion sort - 정렬되지 않은 data둥 가장 작은data를 선택하여 맨 앞부터 순서대로 정렬해 나가는 알고리즘 

def selection_sort(arr):
    for i in range(len(arr) - 1): 
        min_index = i # 최소값 인덱스 0으로 설정 (i가 0부터 시작하기 때문에)
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_index]:
                min_index = j 
        arr[i], arr[min_index] = arr[min_index], arr[i] # 파이썬에서 값 교환해주는 간단한 명령어 
    print(arr)
arr=[43,2,6,8,5,123,63,7,34]
selection_sort(arr)

```

```python
# 버블 정렬(단순 교환법) 
# 인접한 data를 교환하는 처리를 반복한다 

def bubble_sort(arr):
    for i in range(0,len(arr)):
        for j in range(len(arr)-1,i,-1):
            if arr[j-1] > arr[j]:
                arr[j-1], arr[j] = arr[j], arr[j-1]
    return arr
arr=[423,31,423,54,22,36,1]
print(bubble_sort(arr))
```

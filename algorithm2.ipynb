```python
# 최대 값을 찾는 알고리즘 

def get_max(a):
  max_val = a[0]
  for i in range(len(a)):
    if max_val < a[i]:
      max_val = a[i]
  print(max_val)

arr = [12,13,11,14,10]
get_max(arr)
```

```python
# Linear Search 선형 탐색법
# 보관된 data를 맨 앞부터 순서대로 원하는 값을 찾을때 까지 탐색한다
# 탐석 처리는 반복 구조를 사용하여 처리한다 
# 반복 구조에서는 탈출 조건을 잊지 말것 


def linear_search(element, arr):
  for i in range(0,len(arr)): # 리스트의 길이에 따라가  바꿔줄것 
    if arr[i] == element:
      return print(f'{i}번째 요소가 일치') # 함수를 바로 종료시킬 수 있음 
    else:
      i+=1
  return "찾지 못했습니다"
  

arr=[4,2,3,5,1]

linear_search(7,arr)


```

선형 탐색 알고리즘 순서도.jpg

```python
# Binary Search 이진 탐색법
def binary_search(target, data): # 탐색값 , 입력 리스트(배열)
    data.sort() # 이진탐색은 data의 정렬이 되어있다는 조건이 있기 때문에 정렬을 먼저 해준다
    start = 0
    end = len(data) - 1

    while start <= end:
        mid = (start + end) // 2 # 이진탐색을 위한 중간지점
        if data[mid] == target:
            return print(f'{mid}번째 요소가 일치') # 함수를 종료 시켜버릴 수 있음
        elif data[mid] < target:
            start = mid + 1
        else:
            end = mid -1

    return "찾지 못했습니다" # 리턴값을 없다고 할때
arr=[1,4,7,6,9,33,65,37]
binary_search(33,arr)
```

```python
# 예외 처리 (에러 처리, 오류 처리)

# 프로그램 실행 전에 발생하는 오류 - 구문 오류 syntax error
# 프로그램 실행 중에 발생하는 오류 - 런타임 오류 runtime error , 예외 exception

# try except
try: # 오류가 발생하지 않는 한 except로 이동하지 않고 try에서 계속 진행한다 
  num_input=input("정수 입력: ")
  num_a= int(num_input)

  print('원의 반지름: ',num_a)
  print('원의 둘레: ',2*3.14*num_a)
  print('원의 넓이: ',3.14*num_a*num_a)
except: # 오류가 발생하면 except로 넘어간다
  print("정수를 입력하세요")
```

```python
# 숫자만 추출하는 프로그램
list_all = ['52','273','32','스파이','103']

list_num=[]

for i in list_all:
  try:
    float(i)
    list_num.append(i)
  except:
    pass # 에러 발생시 무시하고 넘김 
print(list_num)

try:
  print(list_num[5])
except:
  print("존재하지 않는 인덱스 입니다")
```

```python
try:
  num = int(input("정수를 입력하세요: "))
except:
  print("정수를 입력하지 않았습니다")
else: # 에러가 발생하지 않으면 실행 하지만 사용을 권장하지는 않는다
  print('이상이 없습니다')
finally: # 보통 한 리소스를 종료시킬때 사용
  print('프로그램이 완전히 종료되었습니다')
```

```python
def test(): # 함수의 return은 else보단 강하고 finally보단 약하다 리턴을 만나도 finally는 실행되고 리턴한다
  print('함수의 첫 줄입니다')
  try:
    print('try 실행')
    return
    print('return 아래')
  except:
    print('except 실행')
  else:
    print('else 실행')
  finally:
    print('finally 실행')
  print('함수의 마지막')

test()
```

```python
def count_letter(a): # 대문자와 소문자를 카운트하는 함수 
  cnt_u = 0 # 대문자 개수
  cnt_l = 0 # 소문자 개수
  for i in a:
    if i.isupper(): # 문자열 반복문 돌면서 대문자가 나오면
      cnt_u += 1 # 대문자 개수를 카운트하는 변수를 하나씩 증가
    elif i.islower(): # 문자열 반복문 돌면서 소문자가 나오면
      cnt_l += 1 # 소문자 개수를 카운트하는 변수를 하나씩 증가
  print(f'The number of lowercase characters is :  \n {cnt_l}') # 소문자 개수 출력문 
  print(f'The number of uppercase characters is :  \n {cnt_u}') # 대문자 개수 출력문
def swap_letter(a):  # 대문사 소문자 변환해주는 함수 
  swap_str='' # 변환된 문자열을 출력하기 위한 문자열 선언 
  for i in a:
    if i.isupper():
      swap_str =swap_str + i.lower() # 대문자일 경우 소문자로 변환
    elif i.islower():
      swap_str = swap_str + i.upper() # 소문자일 경우 대문자로 변환
  print(f'swap case: {swap_str}') # 대소문자 변환된 문자열 출력 
for i in range(1,3): # Test Case 2 번실행
  string = input('Enter String:')
  count_letter(string)
  swap_letter(string)
  print('\n') # Test 실행할때 가시성을 위한 간격
```

```python
# 에라토스테네스의 체 두 정수 사이의 소수 개수를 구하기 

def prime(number):
    if number == 1:
        return False
    else:
        for i in range(2, int(number ** 0.5) + 1): # 제곱근한 수 이하의 소수들의 배수들을 제거하고 남은 숫자들이 소수다 --> 극한으로 가면 100% 정확하지는 않다 
            if number % i == 0:
                return False
        return True

M, N = map(int, input().split())
for i in range(M, N + 1):
    if prime(i):
        print(i)
```

```python
# 유클리드의 호제법 최대공약수 구하기 알고리즘
# x,y 의 최대 공약수는 x,r(= x%y)의 최대 공약수 이다 
def gcd(a, b): # Greatest Common Divisor
  if a < b: 
    (a, b) = (b, a) 
  while b !=0: 
    (a, b) = (b, a % b) 
  return a

print(gcd(221,143))
# 선 판정형 구조는 반복 처리에 들어가기 전에 조건식으로 판정을 실시한다
# 후 판정형 구조는 반복 처리 후에 조건식으로 판정을 실시한다

# 최소 공배수 Least Common Multiplier
def lcm(a,b):
  if a < b: 
    (a, b) = (b, a)
  result = (a*b) // gcd(a,b)
  return result

print(lcm(221,143))
```

from django.db import models

class Question(models.Model) : subject = models.CharField(max_length=100) content = models.TextField() create_date = models.DateTimeField()

class Answer(models.Model) : subject = models.ForeignKey(Question, on_delete=models.CASCADE) content = models.TextField() create_date = models.DateTimeField()

Answer 모델 질문에 대한 답. 기존 모델 속성과 연결하기 위해 ForeignKey 를 사용하여 Question모델과 연결한다. on_delete=models.CASCADE의 의미는 Question 이 삭제될 경우 Answer도 함께 삭제해라라는 의미.
